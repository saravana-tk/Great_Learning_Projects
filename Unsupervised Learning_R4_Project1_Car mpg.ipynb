```python
#Import the required libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Read the input dataset
df = pd.read_csv('cars-dataset.csv')
```

```python
#Print the top 5 rows
df.head()
```

```python
#Check for null values
df.isnull().sum()
```

```python
#Check the number of rows and columns
df.shape
```

```python
#Print the info about the dataframe
df.info()
```

```python
#Print the datatype of each of the columns
df.dtypes
```

```python
#Dropping the car name column
df = df.drop(['car name'], axis=1) 
```

```python
#Check if the columns are all numeric
df.applymap(np.isreal)
```

```python
#'hp' feature seem to have ? for some of the rows
df[df['hp'] == "?"]
```

```python
#Replace the 'hp' column with NAN values
df['hp'][df['hp'] == "?"] = np.nan
```

```python
#Conver the 'hp' column to float
df['hp'] = df['hp'].astype(float)
```

```python
df.applymap(np.isreal)
```

```python
#Check for null values again
df.isnull().sum()
```

```python
#Print the 32nd row. It has null value for the 'hp' column
df.iloc[[32]]
```

```python
#Fill the null values with mean
df['hp'].fillna((df['hp'].mean()), inplace=True)
```

```python
#Check for null values
df.isnull().sum()
```

```python
#Print the 32nd row. The null value has been replaced by the mean of the column.
df.iloc[[32]]
```

```python
#Print the pair plot
sns.pairplot(df)
```

```python
#Import the required libraries and find the ideal number of clusters
#Print the last 12 joins and visualize vertically
#2 clusters seem to be optimal
from scipy.cluster.hierarchy import dendrogram, linkage

Z = linkage(df, method='ward')

# calculate full dendrogram
plt.figure(figsize=(25, 15))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('sample index')
plt.ylabel('distance')
dendrogram(
    Z,
    truncate_mode='lastp',  # show only the last p merged clusters
    p=12,  # show only the last p merged clusters
    leaf_rotation=90.,  # rotates the x axis labels
    leaf_font_size=8.,  # font size for the x axis labels
)
plt.show()
```

```python
#Separate the clusters
from scipy.cluster.hierarchy import fcluster
t = 2
clusters = fcluster(Z, t, criterion='maxclust')
clusters
```

```python
#Add the cluster column to the dataframe
df['clusters'] = clusters
df.head(20)
```

```python
#Create Cluster 1
cluster_1 = df[df['clusters']==1]
cluster_1
```

```python
#Create Cluster 2
cluster_2 = df[df['clusters']==2]
cluster_2
```

```python
#Import Linear regression from sklearn
from sklearn.linear_model import LinearRegression
```

```python
#Create 2 data frame with features and target for Cluster 1
cluster_1.drop(['clusters'], axis=1,inplace=True)
X_cluster1 = cluster_1.drop(['mpg'], axis=1)
y_cluster1 = cluster_1['mpg']
```

```python
#Create 2 data frame with features and target for Cluster 2
cluster_2.drop(['clusters'], axis=1,inplace=True)
X_cluster2 = cluster_2.drop(['mpg'], axis=1)
y_cluster2 = cluster_2['mpg']
```

```python
#Print the mpg of cluster 1
y_cluster1
```

```python
linearRegressor = LinearRegression()
```

```python
#Perform linear regression for cluster 1
linearRegressor.fit(X_cluster1, y_cluster1)
```

```python
#Print the coeffiecients
linearRegressor.coef_
```

```python
#Print the coeffiecients in the readable format
coefficients = pd.concat([pd.DataFrame(X_cluster1.columns),pd.DataFrame(np.transpose(linearRegressor.coef_))], axis = 1)
print('Coefficients for Cluster1:') 
coefficients
```

```python
#Perform linear regression for cluster 2
# and Print the coeffiecients in the readable format

linearRegressor.fit(X_cluster2, y_cluster2)
linearRegressor.coef_
coefficients = pd.concat([pd.DataFrame(X_cluster2.columns),pd.DataFrame(np.transpose(linearRegressor.coef_))], axis = 1)
print('Coefficients for Cluster2:') 
coefficients
```
