```python
#Import the StandardScaler function from the library
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
#Split the predictor and target variables
X_train = df.drop(columns = ['status'])
Y_train = df['status']
```

```python
#Import the train test split and split the data into train and test in 80:20 ratio
from sklearn.model_selection import train_test_split

test_size = 0.2 
seed = 7 
x_train, x_test, y_train, y_test = train_test_split(X_train, Y_train, test_size=test_size, random_state=seed)
```

```python
#Scale the training data using Standard Scaler
scaler.fit(x_train)
```

```python
#Scale the test data using Standard Scaler
scaler.fit(x_test)
scaledx_test = scaler.transform(x_test)
```

```python
# loading KNN library and fitting the model
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
# empty list that will hold accuracy scores
ac_scores = []

for k in range(1, 55, 1) :
  model = KNeighborsClassifier(n_neighbors = k)    # initialize the model
  model.fit(scaledx_train, y_train)                      # fit the model
  y_pred = model.predict(x_test)                   # make predictions
  score = accuracy_score(y_test, y_pred)           # get accuracy
  ac_scores.append((score, k))                     # append accuracy score
  
# determining best k
print(ac_scores)
print("The max accuracy and optimal number of neighbors is {}".format(max(ac_scores)))
```

```python
#Plotting the accuracy using pyplot.
import matplotlib.pyplot as plt

score, k = zip(*ac_scores)
plt.plot(k, score)
plt.xlabel('Number of Neighbors K')
plt.ylabel('Accuracy score')
plt.show()
```
